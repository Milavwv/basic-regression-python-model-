```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score

rain_df =pd.read_csv("train.csv")
test_df =pd.read_csv("test.csv")

x_train =train_df["x"].values.reshape(-1,1)
y_train =train_df["y"].values


print(pd.isna(y_train).sum())
mean_value =np.nanmean(y_train)
print(mean_value)
y_train = np.where(np.isnan(y_train),mean_value,y_train)



x_test =test_df["x"].values.reshape(-1,1)
y_test =test_df["y"].values

model=LinearRegression()
model.fit(x_train,y_train)

y_pred =model.predict(x_test)

mse =mean_squared_error(y_test,y_pred)
r2 =r2_score(y_test,y_pred)
print(f"MSE:{mse}")
print(f" R²:{r2}")

plt.scatter(x_test,y_test,color="red",label="real number")
plt.plot(x_test,y_pred,color="blue",label="predict line")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression practice")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.scatter(x_train, y_train, color='blue', label='Train')
plt.scatter(x_test, y_test, color='red', label='Test')
plt.legend()
plt.title("Train vs Test Distribution")
plt.grid(True)
plt.show()
```

```python
plt.hist(x_train, bins=30, alpha=0.5, label="x_train", color='blue')
plt.hist(x_test, bins=30, alpha=0.5, label="x_test", color='red')
plt.legend()
plt.title("x Distribution - Train vs Test")
plt.xlabel("x")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
