```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
X = np.random.rand(500, 1)
Y = 2 * X + 1 + np.random.normal(0, 0.1, (500, 1))

W = np.random.randn(1, 1)
B = np.zeros((1, 1))
learning_rate = 0.01
epochs = 1000
N = X.shape[0]

losses = []
for epoch in range(epochs):
    Y_pred = X @ W + B
    error = Y_pred - Y

    dW = (1/N) * X.T @ error
    dB = (1/N) * np.sum(error)

    W = W - learning_rate * dW
    B = B - learning_rate * dB

    if epoch % 100 == 0:
        loss = (1/(2*N)) * np.sum(error ** 2)
        losses.append(loss)
        print(f"Epoch {epoch}: Loss = {loss:.4f}")

plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', label='Data Points', alpha=0.5)
Y_pred_final = X @ W + B
plt.plot(X, Y_pred_final, color='red', label='Learned Line', linewidth=2)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression with Gradient Descent')
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(np.arange(0, epochs, 100), losses, color='green', marker='o')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss Curve During Training')
plt.show()

print(f"Trained Weight (W): {W}")
print(f"Trained Bias (B): {B}")
```
